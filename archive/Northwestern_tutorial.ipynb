```python
import numpy as np
import jax.random as random

import matplotlib.pyplot as plt

from utils_node import RK_forward_pass_nobias as NODE_nobias, RK_forward_pass as NODE
from utils_node import init_layers, init_layers_nobias
```

```python
key = random.PRNGKey(2023)
layers = [1,5,5,5,1]
params = init_layers(layers, key)

x = np.linspace(0,1).reshape([50,1])
y = NODE(x, params)
```

```python
plt.plot(x,y)
```

```python
for i in range(100):
    key, subkey = random.split(key)
    params = init_layers(layers, key)
    y = NODE(x, params)
    plt.plot(x,y)
```

```python
for i in range(100):
    key, subkey = random.split(key)
    params = init_layers_nobias(layers, key)
    y = NODE_nobias(x, params)
    plt.plot(x,y)
```
