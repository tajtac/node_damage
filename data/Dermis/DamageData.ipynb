```python
import numpy as np
import pandas as pd
import os
```

```python
path = "DamageData/"
os.chdir(path=path)
files = next(os.walk("."))[2]

```

```python
def get_details(filename):
    Pig = filename[1]
    if "Breast" in filename:
        Location = "Breast"
    elif "Belly" in filename:
        Location = "Belly"
    else:
        Location = "ND"
    return [filename.removesuffix("Data.csv"),Pig,Location]
```

```python
Data = []
for filename in files:
    dat = get_details(filename)
    Data.append(dat)
os.chdir("..")
Data = np.array(Data)
DamageData = pd.DataFrame(Data,columns=["Specimen","Pig","Location"])
```

```python
DamageData.to_csv("DamageData.csv",index=False)
```
